```python
# 做整合
from processor.document_processor import DocumentProcessor

def demo_document_processor():
    # 1. 初始化文档处理器
    processor = DocumentProcessor(
        directory_path="files", 
        chunk_size=128,   
        overlap=20        
    )
    
    # 2. 获取目录中文件的统计信息
    stats = processor.get_file_stats()
    print("\n=== 文件统计信息 ===")
    print(f"总文件数: {stats['total_files']}")
    print("文件类型分布:")
    for ext, count in stats["extension_counts"].items():
        print(f"  {ext} ({processor.get_extension_type(ext)}): {count}文件")
    print(f"总文本长度: {stats['total_content_length']}字符")
    print(f"平均文件长度: {stats['average_file_length']:.2f}字符")
    
    # 3. 处理所有支持的文件
    print("\n=== 处理所有文件 ===")
    results = processor.process_directory()
    
    # 4. 显示处理结果
    for result in results:
        print(f"\n文件: {result['filename']}")
        print(f"类型: {processor.get_extension_type(result['extension'])}")
        print(f"内容长度: {result['content_length']}字符")
        
        if result.get("chunks"):
            print(f"分块数量: {result['chunk_count']}")
            print(f"平均分块长度: {result['average_chunk_length']:.2f}字符")
            
            # 打印前两个分块的内容示例
            print("\n分块内容示例:")
            for i, chunk in enumerate(result["chunks"][:2]):
                print(f"\n块 {i+1}:")
                print(''.join(chunk))
                if i >= 1:  # 只显示前两个块
                    print("...")
                    break
        else:
            print(f"分块失败: {result.get('chunk_error', '未知错误')}")

    # 5. 处理特定类型的文件
    print("\n=== 只处理特定类型文件 ===")
    txt_results = processor.process_directory(['.pdf', '.doc'])  # 只处理 txt 和 md 文件
    print(f"处理了 {len(txt_results)} 个文本文件")
    print(txt_results)
    return txt_results
txt_results = demo_document_processor()
```

```python
import faiss
from model_use.embedding_model import init_embedding_model
from sentence_transformers import SentenceTransformer
import pickle
class RetrievalStrategy:
    def __init__(self):
        self.sentence_transformer = SentenceTransformer("models/all-MiniLM-L6-v2")
        self.index = None
        self.doc_store = []  # 存储原始文本，与索引中的向量一一对应

    def build_faiss_index(self, documents):
        """
        documents:就是文本分块的列表
        """
        embeddings = self.sentence_transformer.encode(documents, convert_to_numpy=True)
        dim = embeddings.shape[1]
        self.index = faiss.IndexFlatL2(dim)
        self.index.add(embeddings)
        self.doc_store = documents

        # 可选：保存
        faiss.write_index(self.index, "vector_store/faiss.index")
        with open("vector_store/doc_store.pkl", "wb") as f:
            pickle.dump(self.doc_store, f)

    def load_faiss_index(self):
        self.index = faiss.read_index("vector_store/faiss.index")
        with open("vector_store/doc_store.pkl", "rb") as f:
            self.doc_store = pickle.load(f)
        return self.index,self.doc_store
a = RetrievalStrategy()
a.build_faiss_index(texts)
```

```python
# 嵌入模型测试
from langchain_community.embeddings import HuggingFaceEmbeddings
import numpy as np
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
if device == "cuda":
    print(f"GPU型号: {torch.cuda.get_device_name(0)}")
def init_embedding_model():
    """初始化嵌入模型"""
    return HuggingFaceEmbeddings(
        model_name="models/all-MiniLM-L6-v2",  # 使用本地模型路径
        model_kwargs={'device': device},
        encode_kwargs={'normalize_embeddings': True}
    )

def get_text_embedding(text: str, model):
    """获取文本的嵌入向量"""
    return model.embed_query(text)

def get_text_embeddings(texts: list, model):
    """批量获取多个文本的嵌入向量"""
    return model.embed_documents(texts)

def calculate_similarity(vec1, vec2):
    """计算两个向量的余弦相似度"""
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

def demo():
    # 1. 初始化模型
    model = init_embedding_model()
    
    # 2. 准备示例文本
    text1 = "中国的发展模式强调合作共赢"
    text2 = "中国特色现代化道路追求互利共赢"
    text3 = "人工智能技术快速发展"
    
    # 3. 获取单个文本的嵌入向量
    embedding1 = get_text_embedding(text1, model)
    embedding2 = get_text_embedding(text2, model)
    embedding3 = get_text_embedding(text3, model)
    print(f"\n文本1的嵌入向量维度: {len(embedding1)}")
    print(f"向量前5个值: {embedding1[:5]}")
    print(f"\n文本1的嵌入向量维度: {len(embedding2)}")
    print(f"向量前5个值: {embedding2[:5]}")
    print(f"\n文本1的嵌入向量维度: {len(embedding3)}")
    print(f"向量前5个值: {embedding3[:5]}")    
    # 4. 批量获取多个文本的嵌入向量
    texts = [text1, text2, text3]
    embeddings = get_text_embeddings(texts, model)
    print(f"\n批量处理的向量数量: {len(embeddings)}")
    
    # 5. 计算文本相似度
    sim_1_2 = calculate_similarity(embeddings[0], embeddings[1])
    sim_1_3 = calculate_similarity(embeddings[0], embeddings[2])
    
    print("\n相似度比较:")
    print(f"文本1 vs 文本2 (相似主题): {sim_1_2:.4f}")
    print(f"文本1 vs 文本3 (不同主题): {sim_1_3:.4f}")


demo()
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from langchain_huggingface import HuggingFacePipeline
import torch

def init_deepseek_model():
    model_name = "models/DeepSeek-R1-Distill-Qwen-7B"
    
    # 1. 配置 tokenizer
    tokenizer = AutoTokenizer.from_pretrained(
        model_name,
        trust_remote_code=True,
        use_fast=True  # 使用快速分词器
    )
    
    # 2. 配置模型加载和量化参数
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        device_map="auto",
        trust_remote_code=True,
        use_cache=True,
        torch_dtype=torch.float16,  # 使用 FP16
        quantization_config={
            "load_in_4bit": True,  # 使用 4bit 量化
            "bnb_4bit_compute_dtype": torch.float16,
            "bnb_4bit_use_double_quant": True,  # 使用双重量化进一步节省显存
            "bnb_4bit_quant_type": "nf4"  # 使用 NF4 量化类型
        },
        # 优化配置
        low_cpu_mem_usage=True,
        offload_folder="offload"  # 设置模型权重卸载目录
    )
    
    # 3. 配置生成参数
    pipe = pipeline(
        "text-generation",
        model=model,
        tokenizer=tokenizer,
        max_new_tokens=512,
        temperature=0.7,
        top_p=0.9,
        repetition_penalty=1.1,
        do_sample=True,
        device_map="auto",
        # 添加批处理和缓存配置
        batch_size=1,
        use_cache=True
    )
    
    print("模型加载完成，当前显存占用：")
    print(f"GPU 显存占用: {torch.cuda.memory_allocated()/1024**2:.2f} MB")
    
    return HuggingFacePipeline(pipeline=pipe)
```

```python
llm = init_deepseek_model()

# 测试问题
test_questions = [
    "请简单介绍一下中国式现代化的特点",
    "什么是人类命运共同体？",
]

for question in test_questions:
    print(f"\n问题：{question}")
    response = llm(question)
    print(f"回答：{response}")
    print(f"当前显存占用：{torch.cuda.memory_allocated()/1024**2:.2f} MB")
```

```python
from rank_bm25 import BM25Okapi
import hanlp
import numpy as np

# 加载分词器（建议放外面避免重复加载）
tokenizer = hanlp.load(hanlp.pretrained.tok.COARSE_ELECTRA_SMALL_ZH)

def bm25_retrieval(query, documents, tokenized_documents, k=5):
    """中文 BM25 检索"""
    # 初始化 BM25
    bm25 = BM25Okapi(tokenized_documents)
    
    # 查询分词
    query_tokens = tokenizer(query)
    

    # 计算 BM25 得分
    scores = bm25.get_scores(query_tokens)

    # 取 top-k 得分最高的文档索引
    top_k_idx = np.argsort(scores)[-k:][::-1]

    # 返回对应的文档
    return [documents[i] for i in top_k_idx], [scores[i] for i in top_k_idx]
```

```python
def remove_markdown_formatting(text: str) -> str:
    """去除文本中的Markdown格式字符"""
    # 去除标题格式 (###)
    text = text.replace("### ", "")
    
    # 去除粗体格式 (**文本**)
    import re
    text = re.sub(r'\*\*(.*?)\*\*', r'\1', text)
    
    # 去除斜体格式 (*文本*)
    text = re.sub(r'\*(.*?)\*', r'\1', text)
    
    # 去除代码格式 (`文本`)
    text = re.sub(r'`(.*?)`', r'\1', text)
    
    # 去除链接格式 [文本](链接)
    text = re.sub(r'\[(.*?)\]\(.*?\)', r'\1', text)
    
    # 去除图片格式 ![alt](链接)
    text = re.sub(r'!\[(.*?)\]\(.*?\)', r'\1', text)
    
    # 去除水平线
    text = re.sub(r'\n---+\n', '\n', text)
    
    # 去除引用格式 (> 文本)
    text = re.sub(r'^\s*>\s*(.*?)$', r'\1', text, flags=re.MULTILINE)
    
    # 去除列表标记符号 (- 或 * 或 1. 等)
    text = re.sub(r'^\s*[-*+]\s+', '', text, flags=re.MULTILINE)
    text = re.sub(r'^\s*\d+\.\s+', '', text, flags=re.MULTILINE)
    
    return text

def extract_after_second_think(prediction: str) -> str:
    """提取第二个</think>后的内容"""
    # 查找第二个</think>的位置
    first_think_pos = prediction.find("think>")
    if first_think_pos == -1:
        return prediction  # 没有找到</think>
    
    # 从第一个</think>后开始搜索第二个</think>
    second_think_pos = prediction.find("think>", first_think_pos + 1)
    if second_think_pos == -1:
        return prediction  # 没有找到第二个</think>
    
    # 返回第二个</think>后的内容
    result = prediction[second_think_pos + len("think>"):]
    result = remove_markdown_formatting(result)
    return result.strip()  # 去除前后空白字符

# 使用示例
# prediction = "一些内容<think>思考过程1</think>中间内容<think>思考过程2</think>最终答案"
final_answer = extract_after_second_think(prediction)
print(final_answer)  # 输出: 最终答案
```
